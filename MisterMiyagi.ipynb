```
def talk():
  input = sr.Recognizer()
  with sr>Microphone as source:
    audio = input.listen(source)
    data = ""
    try:
      data = input.recognize_google(audio)
      print("Your question is, " + data)
    except sr.UnknownValueError:
      print("Speak clearly kodomo.")
    return data
```
